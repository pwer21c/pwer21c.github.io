```python
class Student:
    ## variable
    nom_prenom="bang sanghun"
    ## function
    def info(self):
        print("Je m'appelle "+self.nom_prenom )
        print("My name is "+self.nom_prenom)
        
instance1=Student()
instance1.nom_prenom
instance1.info()
```

```python
class student2:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def adresse(self,add):
        self.add=add
        print(self.add)
        
    def yourclass(self,yc):
        self.yc=yc
        print(self.yc)
        
p1=student2("bang sanghun",49)

print(p1.name)
print(p1.age)
p1.adresse("14 rue edouard vaillant 93400 st-ouen")
p1.yourclass("CM1 ou cm2")
```

```python
p2=student2("kim",32)
print(p2.name)
print(p2.age)
```

```python
class student2:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        
    def myfunction(self):
        print("Hello my name is "+self.name)
        
p1=student2("bang",36)
print(p1.name)
print(p1.age)
p1.myfunction()
```
